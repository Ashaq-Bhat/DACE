```python
# prompt: give me graph  for comparison and analysis 5 best ml methods in crop decease detection from lasttest datasets, also give source of datasets

!pip install tensorflow tensorflow-datasets matplotlib seaborn scikit-learn

import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import numpy as np

# --- Dataset Source ---
# The PlantVillage dataset is a widely used resource for crop disease detection.
# It contains images of healthy and diseased plant leaves.
# Source: https://www.plantvillage.org/ (Often available through TensorFlow Datasets)
# Other sources: Kaggle competitions (search for "crop disease detection"), research papers.

# --- Load the Dataset (using PlantVillage as an example) ---
# Note: This dataset loading might take some time.
train_ds = None
test_ds = None
num_classes = None
class_names = None

try:
    # Load the entire train split
    dataset, info = tfds.load('plant_village', split='train', with_info=True)

    # Split the train dataset into train and test
    train_dataset = dataset.take(int(info.splits['train'].num_examples * 0.8))
    test_dataset = dataset.skip(int(info.splits['train'].num_examples * 0.8))


    # Preprocess the data (resize images, normalize, batch, etc.)
    IMG_WIDTH = 128
    IMG_HEIGHT = 128
    BUFFER_SIZE = 1000 # Set a concrete buffer size for shuffling
    BATCH_SIZE = 32

    def preprocess(features):
        image = features['image']
        label = features['label']
        image = tf.image.resize(image, (IMG_WIDTH, IMG_HEIGHT))
        image = tf.cast(image, tf.float32) / 255.0
        return image, label

    train_ds = train_dataset.map(preprocess).shuffle(BUFFER_SIZE).batch(BATCH_SIZE).prefetch(buffer_size=tf.data.AUTOTUNE)
    test_ds = test_dataset.map(preprocess).batch(BATCH_SIZE).prefetch(buffer_size=tf.data.AUTOTUNE)

    num_classes = info.features['label'].num_classes
    class_names = info.features['label'].names
    print(f"Number of classes: {num_classes}")
    print(f"Class names: {class_names}")

except tf.errors.NotFoundError:
    print("PlantVillage dataset not found in TensorFlow Datasets.")
    print("You may need to download it manually or use a different dataset source.")
    print("Setting num_classes to a default value. Model training will not be possible without a dataset.")
    num_classes = 10 # Set a default value
except Exception as e:
    print(f"An error occurred during dataset loading or splitting: {e}")
    print("Setting num_classes to a default value. Model training will not be possible without a dataset.")
    num_classes = 10 # Set a default value


# --- Define ML Models ---
# Here we define the architecture of 5 different ML models.
# For a real analysis, you would implement or load pre-trained versions
# of models suitable for image classification.

def build_simple_cnn(input_shape, num_classes):
    model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(num_classes, activation='softmax')
    ])
    return model

def build_cnn_with_dropout(input_shape, num_classes):
    model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Dropout(0.25),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Dropout(0.25),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(num_classes, activation='softmax')
    ])
    return model

# Add more model definitions here (e.g., a transfer learning model based on VGG16 or ResNet)
def build_vgg16_transfer(input_shape, num_classes):
    base_model = tf.keras.applications.VGG16(input_shape=input_shape,
                                               include_top=False,
                                               weights='imagenet')
    base_model.trainable = False # Freeze the convolutional base

    model = tf.keras.Sequential([
        base_model,
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dense(num_classes, activation='softmax')
    ])
    return model

def build_resnet50_transfer(input_shape, num_classes):
    base_model = tf.keras.applications.ResNet50V2(input_shape=input_shape,
                                                include_top=False,
                                                weights='imagenet')
    base_model.trainable = False # Freeze the convolutional base

    model = tf.keras.Sequential([
        base_model,
        tf.keras.layers.GlobalAveragePooling2D(),
        tf.keras.layers.Dense(num_classes, activation='softmax')
    ])
    return model

# Example of a simpler non-CNN model (might not be suitable for large image datasets)
# from sklearn.ensemble import RandomForestClassifier
# from sklearn.preprocessing import StandardScaler
# def train_random_forest(train_ds, test_ds):
#     # For Random Forest, you would need to extract features from images
#     # and flatten them. This is more complex for large datasets.
#     # This part is commented out as it requires significant feature engineering.
#     print("Random Forest training requires manual feature extraction and flattening.")
#     return None, None # Return None for metrics if not implemented

# --- Prepare Models for Comparison ---
input_shape = (IMG_WIDTH, IMG_HEIGHT, 3)

models = {
    'Simple CNN': build_simple_cnn(input_shape, num_classes),
    'CNN with Dropout': build_cnn_with_dropout(input_shape, num_classes),
    'VGG16 Transfer Learning': build_vgg16_transfer(input_shape, num_classes),
    'ResNet50 Transfer Learning': build_resnet50_transfer(input_shape, num_classes),
    # Add more models here. You could try:
    # 'InceptionV3 Transfer Learning': build_inceptionv3_transfer(input_shape, num_classes),
    # Note: Add the corresponding build function for each model.
}

# --- Train and Evaluate Models ---
results = []

# Only proceed with training and evaluation if the dataset was loaded successfully
if train_ds and test_ds and num_classes is not None:
    for model_name, model in models.items():
        print(f"\n--- Training {model_name} ---")

        # Compile the model
        model.compile(optimizer='adam',
                      loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                      metrics=['accuracy'])

        # Train the model (Adjust epochs based on your needs and resources)
        epochs = 5 # Keep epochs low for demonstration. Increase for better performance.
        history = model.fit(train_ds, epochs=epochs, validation_data=test_ds)

        # Evaluate the model on the test set
        print(f"\n--- Evaluating {model_name} ---")
        loss, accuracy = model.evaluate(test_ds)
        print(f"Test Accuracy: {accuracy:.4f}")

        # Get predictions for precision, recall, and F1-score
        y_true = []
        y_pred = []
        for images, labels in test_ds:
            predictions = model.predict(images)
            predicted_labels = tf.argmax(predictions, axis=1).numpy()
            y_true.extend(labels.numpy())
            y_pred.extend(predicted_labels)

        precision = precision_score(y_true, y_pred, average='weighted') # Use 'weighted' for multi-class
        recall = recall_score(y_true, y_pred, average='weighted')
        f1 = f1_score(y_true, y_pred, average='weighted')

        results.append({
            'Model': model_name,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1-score': f1
        })

    # --- Create a DataFrame for Results ---
    results_df = pd.DataFrame(results)
    print("\n--- Model Performance Summary ---")
    print(results_df)

    # --- Generate Comparison Graph ---
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Model', y='Accuracy', data=results_df)
    plt.title('Accuracy Comparison of ML Methods for Crop Disease Detection')
    plt.ylabel('Accuracy')
    plt.ylim(0, 1) # Accuracy is between 0 and 1
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # You can also plot other metrics:
    plt.figure(figsize=(12, 6))
    results_melted = results_df.melt(id_vars=['Model'], value_vars=['Precision', 'Recall', 'F1-score'], var_name='Metric', value_name='Score')
    sns.barplot(x='Model', y='Score', hue='Metric', data=results_melted)
    plt.title('Precision, Recall, and F1-score Comparison')
    plt.ylabel('Score')
    plt.ylim(0, 1)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Metric')
    plt.tight_layout()
    plt.show()
else:
    print("Dataset not loaded successfully. Skipping model training and evaluation.")
```
